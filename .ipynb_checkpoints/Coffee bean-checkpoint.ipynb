```python
df_order = pd.merge(left=order, right=customer[['Customer ID', 'Loyalty Card']], on='Customer ID')
df_order
```

```python
product.head()
```

```python
df_total = pd.merge(left=df_order, right=product, on='Product ID')
df_total
```

```python
df_total['Quantity'].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
%matplotlib inline

sns.countplot(data = df_total, x=df_total.Quantity)
plt.xlabel('Quantity')
plt.ylabel('Count of Customers')
plt.title('Quantity of the coffee product ordered')
plt.show()
```

* Robusta, Exc, Arabica, Lib Coffee
* Ïª§Ìîº Ï¢ÖÎ•òÏóê Îî∞Î•∏ Íµ¨Îß§ ÏàòÎüâ Ï∞®Ïù¥ ÌôïÏù∏

```python
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
%matplotlib inline

plt.figure(figsize=(15, 6))
sns.countplot(data = df_total, x=df_total.Quantity, hue ='Coffee Type')
plt.xlabel('Quantity')
plt.ylabel('Count of Customers')
plt.title('Quantity of the coffee product ordered')
plt.legend(bbox_to_anchor = (1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
