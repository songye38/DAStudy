```python
df_total['Total_Profit'] = df_total['Quantity'] * df_total['Profit']
df_total
```

### Profit in Quarter

```python
df_profit = df_total.groupby(['Quarter'], as_index=False).agg({"Total_Profit" : "sum"})
df_profit
```

```python
df_profit['Quarter'] = df_profit['Quarter'].astype(str)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_profit, x='Quarter', y='Total_Profit')
plt.title('Sales Recording in Quarter')
plt.show()
```

### Unit Price in Quarter

```python
df_total["Total_Unit_Price"] = df_total['Quantity'] * df_total["Unit Price"]
df_total
```

```python
df_unitprice = df_total.groupby('Quarter', as_index=False).agg({"Total_Unit_Price" : "sum"})
df_unitprice
```

```python
df_unitprice['Quarter'] = df_unitprice['Quarter'].astype(str)
```

```python
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_unitprice, x='Quarter', y='Total_Unit_Price')
plt.title('Unit Price Recording in Quarter')
plt.show()
```

#### barplot 

```python
plt.subplots(figsize=(20, 12))
sns.barplot(data=df_coffee_type, x="Quarter", y="Total_Quantity", hue="Coffee Type", palette="pastel")
plt.title("Quarterly consumption of Coffee_type (Beans)")
plt.show()
```

#### lineplot
* relplotÏúºÎ°ú Í∑∏Î¶¨Î†§Í≥† ÌñàÏúºÎÇò, xtickslabelÏù¥ Ï°∞Ï†ïÏù¥ Í≥ÑÏÜçÌï¥ÎèÑ ÏïàÎêòÏÑú Î≥ÄÍ≤Ω

```python
df_coffee_type["Quarter"] = df_coffee_type["Quarter"].astype(str)
```

```python
df_coffee_type.head()
```

```python
fig, axes = plt.subplots(2, 2, sharey=True ,figsize=(14, 8))

sns.lineplot(data = df_coffee_type[df_coffee_type["Coffee Type"] == "Ara"], x="Quarter", y="Total_Quantity", ax=axes[0, 0])
axes[0, 0].xaxis.set_major_locator(plt.FixedLocator(axes[0, 0].get_xticks()))
axes[0, 0].set_xticklabels(axes[0, 0].get_xticklabels(), rotation=30)
axes[0, 0].set_title("Coffee_Type = Ara")

sns.lineplot(data = df_coffee_type[df_coffee_type["Coffee Type"] == "Exc"], x="Quarter", y="Total_Quantity", ax=axes[0, 1])
axes[0, 1].xaxis.set_major_locator(plt.FixedLocator(axes[0, 0].get_xticks()))
axes[0, 1].set_xticklabels(axes[0, 0].get_xticklabels(), rotation=30)
axes[0, 1].set_title("Coffee_Type = Exc")

sns.lineplot(data = df_coffee_type[df_coffee_type["Coffee Type"] == "Lib"], x="Quarter", y="Total_Quantity", ax=axes[1, 0])
axes[1, 0].xaxis.set_major_locator(plt.FixedLocator(axes[0, 0].get_xticks()))
axes[1, 0].set_xticklabels(axes[0, 0].get_xticklabels(), rotation=30)
axes[1, 0].set_title("Coffee_Type = Lib")

sns.lineplot(data = df_coffee_type[df_coffee_type["Coffee Type"] == "Rob"], x="Quarter", y="Total_Quantity", ax=axes[1, 1])
axes[1, 1].xaxis.set_major_locator(plt.FixedLocator(axes[0, 0].get_xticks()))
axes[1, 1].set_xticklabels(axes[0, 0].get_xticklabels(), rotation=30)
axes[1, 1].set_title("Coffee_Type = Rob")

plt.subplots_adjust(hspace=0.5)

plt.show()
```
