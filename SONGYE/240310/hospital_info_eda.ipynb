```python
import os
import requests
from dotenv import load_dotenv
import pandas as pd
import xml.etree.ElementTree as ET

# .env 파일 로드
load_dotenv()

# 인증키 가져오기
encoding_key = os.getenv("HOSPITAL_DECODING_KEY")

# API 엔드포인트
url = 'https://apis.data.go.kr/B551182/hospInfoServicev2/getHospBasisList'
params = {
    'serviceKey': encoding_key,
    'sidoCd':'310000'
    #'sgguCd':'110002'
}

# API 호출
response = requests.get(url, params=params)

# XML 데이터 확인
#print(response.text)

# XML 데이터 파싱
root = ET.fromstring(response.text)

total_count = int(root.find('.//totalCount').text)

# 결과 출력
print("총 데이터 수:", total_count)

# 데이터 추출
data = []
for item in root.iter('item'):
    hospital = {}
    for child in item:
        hospital[child.tag] = child.text
    data.append(hospital)

# DataFrame 생성
df = pd.DataFrame(data)

# 결과 출력
print(df)
```
