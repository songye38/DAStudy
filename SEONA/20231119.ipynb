```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

file_path = './datas/beers.csv'


df_beers = pd.read_csv(file_path)
```

```python
df_beers
```

```python
df_beers.drop(columns=['ibu'], inplace=True)
```

```python
df_beers
```

```python
df_beers.info()
```

```python
df_beers.dropna(inplace=True)
```

```python
df_beers.info()
```

```python
abv_numbers = df_beers['abv'].value_counts().sort_index()

abv_numbers.plot(kind='bar', figsize=(15, 6))
plt.title('abv')
plt.xlabel('abv')
plt.ylabel('numbers')
plt.show()
```

```python
style_numbers = df_beers['style'].value_counts().sort_index()
style_numbers.plot(kind='bar', figsize=(15, 6))
plt.title('style')
plt.xlabel('style')
plt.ylabel('numbers')
plt.show()
```

```python
style_numbers = df_beers['style'].value_counts()
style_numbers = style_numbers.head(10)
style_numbers.plot(kind='bar', figsize=(15, 6))
plt.title('style')
plt.xlabel('style')
plt.ylabel('numbers')
plt.show()
```
