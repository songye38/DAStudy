```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

file_path = 'C:/Users/82108/Documents/t스터디/archive/Raw Data.xlsx'

# 첫 번째 시트 불러오기
df_orders = pd.read_excel(file_path, sheet_name='orders')

# 두 번째 시트 불러오기
df_customers = pd.read_excel(file_path, sheet_name='customers')

# 세 번째 시트 불러오기
df_products = pd.read_excel(file_path, sheet_name='products')
```

```python
df_orders = df_orders.drop(columns=['Customer Name', 'Email', 'Country', 'Coffee Type', 'Roast Type', 'Size', 'Unit Price', 'Sales'])
df_orders
```

```python
#df_orders['Order Year'], df_orders['Order Month'], df_orders['Order Day'] = df_orders['Order Date'].str.split('-').str
#날짜타입 아니고 문자열타입일때
```

```python
df_orders['Order Year'] = df_orders['Order Date'].dt.year
df_orders['Order Month'] = df_orders['Order Date'].dt.month
df_orders['Order Day'] = df_orders['Order Date'].dt.day
```

```python
df_orders
```

```python
yearly_data = {}
for year, data in df_orders.groupby('Order Year'):
    yearly_data[year] = data
```

```python
yearly_data
```

```python
for year, df in yearly_data.items():
    plt.plot(df['Order Montsh'], df['Sales'], label=str(year))
```

```python
# 'Order Date' 열에서 연도별 주문 수를 계산
yearly_orders = df_orders['Order Date'].dt.year.value_counts().sort_index()

# 년도별 주문 수 그래프
yearly_orders.plot(kind='bar', figsize=(10, 6))
plt.title('Yearly Orders')
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# 'Order Date' 열에서 연도별 주문 수를 계산
monthly_orders = df_orders['Order Date'].dt.month.value_counts().sort_index()

# 년도별 주문 수 그래프
monthly_orders.plot(kind='bar', figsize=(10, 6))
plt.title('Monthly Orders')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# 'Order Date' 열에서 연도별 주문 수를 계산
product_orders = df_orders['Product ID'].value_counts()
product_orders = product_orders.head(10)
```

```python

# 년도별 주문 수 그래프
product_orders.plot(kind='barh', figsize=(10, 6))
plt.title('Products Orders')
plt.xlabel('Product')
plt.ylabel('Number of Orders')
plt.show()
```

```python
df_customers = df_customers.drop(columns=['Email', 'Phone Number', 'Address Line 1', 'Postcode'])
```

```python
df_customers
```

```python
country_orders = df_customers['Country'].value_counts().sort_index()

country_orders.plot(kind='bar', figsize=(10, 6))
plt.title('Country Orders')
plt.xlabel('Country')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# 'Country' 열이 'United States'인 데이터만 필터링
us_orders = df_customers[df_customers['Country'] == 'United States']

us_orders
```

```python
us_city_orders = us_orders['City'].value_counts()
top_5_cities = us_city_orders.head(10)
top_5_cities.plot(kind='bar', figsize=(10, 6))
plt.title('US orders')
plt.xlabel('City')
plt.ylabel('Order Number')
plt.show()
```
